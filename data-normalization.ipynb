```python
# Importing necessary libraries
import pandas as pd              # Pandas for data manipulation and analysis
import seaborn as sns            # Seaborn for data visualization
from sklearn import datasets     # Import datasets from scikit-learn for example data
from sklearn.preprocessing import MinMaxScaler  # Import MinMaxScaler for data normalization
import matplotlib.pyplot as plt  # Matplotlib's pyplot for plotting
```

```python
# Load the Iris dataset from scikit-learn
iris = datasets.load_iris()
# Create a pandas DataFrame from the Iris dataset for easier data manipulation
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
# Create pairplot for the DataFrame before normalization to visualize the distributions and relationships
sns.pairplot(df)
# Save the pairplot to an image file named 'original_data.png'
plt.savefig('original_data.png')
```

```python
# Initialize a MinMaxScaler object for normalization
scaler = MinMaxScaler()
# Fit the scaler to the data and transform it, then create a new DataFrame with normalized data
df_normalized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
# Create a pairplot for the normalized DataFrame to visualize the distributions and relationships after normalization
sns.pairplot(df_normalized)
# Save the pairplot of the normalized data to an image file named 'normalized_data.png'
plt.savefig('normalized_data.png')  
# Display the pairplot of the normalized data inline in the Jupyter Notebook
plt.show()  
```
